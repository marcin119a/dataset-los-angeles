```python
from google.cloud import bigquery

client = bigquery.Client()
```

```python
import pandas as pd


gene_list = pd.read_csv('/kaggle/input/oncokb-cancer-gene-list/cancerGeneList.tsv', sep='\t')
gene_list.head()
```

```python
import pandas as pd 
df = pd.read_csv('https://raw.githubusercontent.com/marcin119a/data/refs/heads/main/bq-metadata.csv')
```

```python
df.head()
```

```python
datasets_expression = df[(df['Full ID'].str.contains('RNAseq_hg38_gdc_current')) | (df['Full ID'].str.contains('Expression'))]
```

```python
datasets_mutations = df[(df['Full ID'].str.contains('somatic')) & (df['Full ID'].str.contains('current'))]
```

```python
merged_df = pd.merge(datasets_expression, datasets_mutations, on='BQ Dataset', suffixes=('_expr', '_mut'))
```

```python
filtered_df = merged_df[merged_df['BQ Table_expr'] != 'miRNAseq_hg38_gdc_current']
```

```python
filtered_df['Fields_expr'].iloc[0]
```

```python
genes = ', '.join(f'"{gene}"' for gene in gene_list['Hugo Symbol'].unique())
```

```python
# Tworzenie jednego zapytania z UNION ALL dla wszystkich datasetów
union_queries = []
for dataset in filtered_df['Full ID_expr']:
    union_queries.append(f"""
    SELECT gene_name, case_barcode, tpm_unstranded, primary_site
    FROM {dataset}
    WHERE gene_name in ({genes})
    """)

# Łączenie wszystkich zapytań w jedno z agregacją
query = f"""
SELECT 
    case_barcode,
    ARRAY_AGG(gene_name ORDER BY gene_name) as gene_name,
    ARRAY_AGG(tpm_unstranded ORDER BY gene_name) as tpm_unstranded,
    ANY_VALUE(primary_site) as primary_site
FROM (
    {' UNION ALL '.join(union_queries)}
)
GROUP BY case_barcode
ORDER BY case_barcode
"""

query_job = client.query(query)
df_expressions = query_job.to_dataframe()
```

```python
data_mutations = []
for dataset in filtered_df['Full ID_mut']:
        query = f"""
        SELECT
            m.case_barcode,
            m.Hugo_Symbol,
            m.Start_position,
            m.End_position,
            m.Chromosome,
            m.Variant_Type,
            m.Reference_Allele,
            m.Tumor_Seq_Allele2,
            m.Tumor_Seq_Allele1,
            m.Gene,
            m.Exon,
            m.Strand,
            m.primary_site
        FROM
        {dataset} AS m
        WHERE m.Hugo_Symbol in ({genes})
        ORDER BY m.Hugo_Symbol         
        ;
        """
 
        try: 
            query_job = client.query(query)
            result_df_mutations = query_job.to_dataframe()
            result_df_mutations['dataset'] = dataset
            data_mutations.append(result_df_mutations)
        except Exception as e:
            print(e)
            print(dataset)
```
